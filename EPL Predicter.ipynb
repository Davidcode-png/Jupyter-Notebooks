```python
def rolling_average(group,cols,new_cols):
    group = group.sort_values('date')
    rolling_stats = group[cols].rolling(3,closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ['gf','ga','sh','sot','dist','fk','pk','pkatt']
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_average(group,cols,new_cols)
```

```python
match_rolling = match_df.groupby('team').apply(lambda x: rolling_average(x,cols,new_cols))
```

```python
match_rolling
```

```python
match_rolling = match_rolling.droplevel('team')
```

```python
match_rolling
```

```python
match_rolling.index = range(match_rolling.shape[0])
```

```python
match_rolling
```

```python
from sklearn.metrics import precision_score
```

```python
def makepredictions(data,predictors):
    train = data[data['date'] < "2022-01-01"]
    test = data[data['date'] > "2022-01-01"]
    rf.fit(train[predictors], train["target"])
    prediction = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test['target'],predictions=prediction))
    precision = precision_score(test['target'],prediction)
    return combined,precision
```

```python
combined, precision = makepredictions(match_rolling,predictors+new_cols)
```
